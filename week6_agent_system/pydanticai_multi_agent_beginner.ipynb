# Beginner Multi-Agent App with PydanticAI + Groq

This notebook shows a **simple multi-agent workflow** using:

- **PydanticAI**
- **Groq API**
- Multiple agents
- `.env` file for API key loading

An agent is an AI assistant designed for a specific task.

The example workflow:

1. A **Research Agent** gathers information
2. A **Writer Agent** creates a final response

---


````python
# Load environment variables
from dotenv import load_dotenv
import os

load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")

if groq_api_key:
    print("✅ GROQ_API_KEY loaded successfully")
else:
    print("❌ GROQ_API_KEY not found")
````

## Import PydanticAI

````python
from pydantic_ai import Agent
````

## Create the Research Agent

This agent gathers short research notes.


````python
writer_agent = Agent(
    model='groq:llama-3.3-70b-versatile',
    system_prompt=(
        "You are a helpful writer. "
        "Turn notes into simple beginner-friendly explanations."
    )
)

print("✅ Writer agent ready")
````

## Create the Writer Agent

This agent converts research into a clean final answer.


````python

research_agent = Agent(
    model='groq:llama-3.3-70b-versatile',
    system_prompt=(
        "You are a research assistant. "
        "Give short and beginner-friendly research notes."
    )
)

print("✅ Research agent ready")
````

## Multi-Agent Workflow

The research agent works first.

Then the writer agent uses the research result.


````python

# Topic
topic = "What is Artificial Intelligence?"


# Main async workflow
async def main():

    # Step 1: Research
    research_result = await research_agent.run(
        f"Give 5 short bullet points about: {topic}"
    )

    print("\n=== Research Notes ===")
    print(research_result.output)

    # Step 2: Writer
    final_result = await writer_agent.run(
        f"""
        Using these research notes:

        {research_result.output}

        Write a simple explanation for beginners.
        """
    )

    print("\n=== Final Answer ===")
    print(final_result.output)


# Run everything
await main()
````
